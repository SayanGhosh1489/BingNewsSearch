```python
import os
from datetime import datetime
from dotenv import load_dotenv
import requests
import time
from bs4 import BeautifulSoup
import json
from pymongo import MongoClient
from pprint import pprint
from bson import ObjectId
```

```python
# mongo connection
def mongo_connection(host):
    """
    Use the fuction for MongoDB connection

    return: Connection object to the host
    """

    mongo = MongoClient(host)
    return mongo
```

```python
host = 'mongodb://localhost:27017'
mongo = mongo_connection(host=host)
```

```python
def connect_create_db(db_name,collection_name):
    """
    create or connect to the db and collections

    return mongo DB and collection object
        """
    database = mongo[db_name]
    collection = database[collection_name]
    return database, collection
```
